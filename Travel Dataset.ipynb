### 1.1 Seperating numerical and categorical data

#### 1.1.1 Numerical dataset

```python
# 1. Getting Numerical features from the dataset
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'object']

# 2. Creating a Numerical DataFrame
data_numerical = df[numerical_features]
data_numerical.head()
```

#### 1.1.1.1 Discrete Numerical features

```python
discrete_features = [feature for feature in df.columns]
discrete_features
```

#### 1.1.1.2 Continuous Numerical features

```python
continuous_features = [feature for feature in df.columns]
continuous_features 
```

### 1.1.2 Categorical Dataset

```python
# 1. Getting Categorical features from dataset
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object']

# 2. Creating categorical dataframe
data_categorical = df[categorical_features]
data_categorical.head()
```

```python
# Getting unique categories for each features in categorical features

for features in categorical_features:
    print("The features is '{}' and number of categories are '{}'".format(features,len(df[features].unique())))
```

### 1.3 Missing Values

```python
#Getting features with missing value

feature_nan = [feature for feature in df.columns if df[feature].isnull().sum()]
feature_nan
```

```python
# Getting count of missing values for individual features

for feature in feature_nan:
    print("{} has {} no. of missing values".format(feature, df[feature].isnull().sum()))
```

```python
# Getting percentage of missing values for individual features

for feature in feature_nan:
    print("{} has {} no. of missing values".format(feature, np.round(df[feature].isnull().sum())/df[feature].count()*100))
```
